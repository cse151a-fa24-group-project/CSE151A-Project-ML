```python
import os
import zipfile
import gdown
```

```python
# List of file IDs (from google drive that stores our dataset)
file_ids = [
    '1NPUlNQ72xjnu0EYWWbThB72UFSsTtrDL'
]

# Folder to store downloaded ZIP files
zip_folder = 'zip_files'
os.makedirs(zip_folder, exist_ok=True)

# Folder to extract all ZIP files
extract_to = 'extracted_images'
os.makedirs(extract_to, exist_ok=True)

# download and extract all ZIP files from our google drive
for i, file_id in enumerate(file_ids, 1):
    zip_path = os.path.join(zip_folder, f'file_{i}.zip') # each ZIP file will be stored as file_#.zip

    # downloading file from google drive
    gdown.download(f'https://drive.google.com/uc?id={file_id}', zip_path, quiet=False)

    # Extract the downloaded ZIP file
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)

print(f"Done, stored in {extract_to}")
```

```python
from PIL import Image
import os

# Paths to your image directory and output directory
input_directory = "/content/extracted_images/S3E10/With_Peter"
output_directory = "/content/drive/MyDrive/Data/S3E10_resized/With_Peter"

# Create the output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Loop through all images in the input directory
for filename in os.listdir(input_directory):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):  # Check for valid image extensions
        img_path = os.path.join(input_directory, filename)
        img = Image.open(img_path)

        # Resize the image to 320x240
        img_resized = img.resize((320, 240), Image.LANCZOS)  # Use Image.ANTIALIAS for better quality

        # Save the resized image to the output directory
        output_path = os.path.join(output_directory, filename)
        img_resized.save(output_path)

        print(f"Resized and saved {filename} to {output_directory}")
```
